```python
Transactiondf=SourceDataFrame[['Sub-Category','Customer ID','Sales']]
Unique_Transactiondf = Transactiondf.drop_duplicates(keep='first')
Transactiondf=Unique_Transactiondf
Transactiondf_Sum=Transactiondf.groupby(['Sub-Category','Customer ID'])['Sales'].sum().reset_index()
Transactiondf_Sum['Sales']=Transactiondf_Sum['Sales'].round().astype(int)
Transactiondf_Sum=Transactiondf_Sum.rename(columns={'Customer ID':'CustId','Sub-Category':'CatId'})
Transactiondf_Sum=Transactiondf_Sum.pivot(index='CustId',columns='CatId',values='Sales')
Transactiondf_Sum=Transactiondf_Sum.fillna(0)
```
