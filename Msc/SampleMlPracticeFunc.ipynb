```python
from sklearn.cluster import DBSCAN
import numpy as np

# Your 1D dataset (replace this with your own data)
# Example: X = np.array([1, 2, 3, 4, 111])
X = np.array([1, 2, 3, 4, 7,111])  # Example data

# Reshape to make it a 2D array (required by scikit-learn)
X = X.reshape(-1, 1)

# Fit DBSCAN
epsilon = 10  # Adjust as needed
min_samples = 2  # Adjust as needed
dbscan = DBSCAN(eps=epsilon, min_samples=min_samples)
dbscan.fit(X)

# Extract anomalies (noise points)
anomaly_mask = dbscan.labels_ == -1  # Anomalies are labeled as -1 by DBSCAN
anomalies = X[anomaly_mask]

print("Anomalies:", anomalies)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf

# Generate example temperature data
np.random.seed(0)
months = pd.date_range(start='2020-01-01', periods=60, freq='M')
temperature = np.random.normal(loc=20, scale=5, size=60) + np.sin(np.arange(60)/6)

# Create DataFrame
df = pd.DataFrame({'Date': months, 'Temperature': temperature})
df.set_index('Date', inplace=True)

# Plot ACF
plot_acf(df['Temperature'], lags=20)
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.title('Autocorrelation Function (ACF) Plot for Monthly Temperature Data')
plt.show()
```
